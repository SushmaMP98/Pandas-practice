```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize = (8,5),facecolor ="palegreen")
plt.scatter(df.wt,df.mpg)
plt.xlabel("mpg")
plt.ylabel("wt")
plt.title("corelation plot")
```

```python
plt.figure(figsize = (12,5),facecolor ="palegreen")
plt.hist(df["hp"],facecolor= "black",edgecolor = "red",bins= 8)
```

```python
plt.figure(figsize = (12,5),facecolor ="skyblue")

plt.boxplot(df["hp"],vert = False)
plt.ylabel("mpg",size = 15)
plt.xlabel("boxplot",size = 10)
plt.title("boxplot",size = 20)
```

```python
plt.figure(figsize = (12,5),facecolor ="palegreen")
plt.hist(df["mpg"],facecolor = "black",edgecolor = "red",bins = 5)
plt.xlabel("mpg")
```

```python
#violin plot
plt.figure(figsize = (12,5),facecolor ="palegreen")
plt.violinplot(df["hp"],vert = False)
plt.xlabel("hp")
```
